# Ejercicio 2

```python
Valores= [200, 225, 232, 221, 243, 256, 255]
def retorno_semanal(cantidad_acciones, p_ganancia, p_no_ganancia):
  # Calcular la cantidad invertida cada mes
  Inversion= [x*cantidad_acciones for x in Valores] # Variable local
  # Ahora calculamos las posibles ganancias cada dia
  Ganancias=[y*(0.15)  for y in Inversion]
  # En el resto de los casos el valor ganado es 0
  Perdidas=[y*(-0.18)  for y in Inversion]
  # Ahora calculamos el valor esperado
  Valor_esperado= [(x*p_ganancia + y*p_no_ganancia) for x,y in zip(Ganancias,Perdidas)]
  return Valor_esperado
retorno_semanal(20,p_ganancia=0.56, p_no_ganancia=0.44)
```

```python
# La suma total luego de una semana es:
sum(retorno_semanal(20,p_ganancia=0.56, p_no_ganancia=0.44))
```

# Ejercicio 3

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/JJTorresDS/stocks-ds-edu/main/stocks.csv'
df = pd.read_csv(url, index_col=0)
df.head()
```

```python
for x in df.columns:
  # Filtrar la columna
  col=df[x]; 
  # Calculo de estadisticas
  media= col.mean(); std= col.std(); var=col.var();
  print('Accion:', x, 'Media: ',media,'Desviacion: ', std, 'Varianza: ', var)
  print('---------------------')
```

```python
for x in df.columns:
  # Filtrar la columna
  col=df[x]; 
  # Calculo de estadisticas
  max= col.max(); min= col.min();
  print('Accion:', x, 'Maximo: ',max,'Minimo: ', min)
  print('---------------------')
```
